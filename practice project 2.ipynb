```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset1/master/mushrooms.csv")
df.head()
```

```python
#All the variables are in string format
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
```

```python
for col in df.columns:
     df[col]=lbl.fit_transform(df[col])
```

```python
# check the data after label encoding
df.head()
```

```python
# split the x and y variables
y=df['class']
x=df.iloc[:,1:23]
```

```python
# check shape of new variables
x.shape
```

```python
y.shape
```

```python
#check data
x.head
```

```python
y.head
```

```python
#first normalise the data using StandardScalar 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#see the Standardised data
print(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
# plot a scrre plot of the principal components
pltfigure(figsize=(16,11))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.grid()
```

```python
new_pca = PCA(n_components=17)
```

```python
x_new = new_pca.fit_transform(x)
```

```python
#using KMeans to plot the clusters.
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=2)
```

```python
k_means.fit_predict(x_new)
```

```python
#plot the clusters
colors = ['r','g']
for i in range(len(x_new)):
    plt.scatter(x_new[i][0], x_new[i][1], c=colors[k_means.labels_[i]], s=10)
  plt.show()
```

```python
x_new.shape
```

```python
#seperate the train and test data
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_tarin, y_test = train_test_split(x_new, y, test_size = 0.25, random_state = 6)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#using logistic regression to build the first model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_predut =lr.predict(x_test)
```
